**LECTURE ASSIGNMENT**

---


Plot using sns.relplot, line plot, box plot on fmri data set


```python
df=sns.load_dataset('fmri')
```

```python
df.head(5)
```

***RELPLOT*** 

---



```python
sns.relplot(x="event", y="region", data=df)
```

```python
sns.relplot(x="timepoint", y="signal",col="region", data=df, height=7)
```

```python
sns.relplot(x="timepoint", y="signal",col="event", data=df, height=7)
```

***LINE PLOT***


---



```python
sns.lineplot(x="timepoint", y="signal", data=df)
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=df)
```

```python
sns.relplot(x="timepoint", y="signal", col="subject",  col_wrap=3, kind="line",data=df)
```

```python
sns.relplot( x="timepoint",y="signal", size="event",hue="region",style="region",kind="line",data=df,height=7 )
```

***BOX PLOT***

---



```python
sns.boxplot(x ="timepoint",y ="signal",hue ="region",data =df)
```

```python
sns.boxplot(x="timepoint",y="signal",data=df)
```
